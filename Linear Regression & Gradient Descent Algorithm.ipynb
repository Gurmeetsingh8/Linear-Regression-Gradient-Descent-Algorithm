```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### data -X =acidity of milk
#y- Density of milk
def readData(filename):
    df=pd.read_csv(filename)
    return df.values
#df.values convert in array
x=readData('./linearX.csv')
y=readData('./linearY.csv')
#print(x)
#print(x.shape)
# we don't want shape as (99,1) so we reshaped it
# reshape is used for convert 2d array into 1d array
x=x.reshape((99,))
y=y.reshape((99,))
plt.scatter(x,y)
plt.show()
```

```python
#Normaliation:- mean =0 and standard deviation =1
x=x-x.mean()
plt.scatter(x,y)
plt.show()
X=x
Y=y
```

### Algorithm-Linear Regression !

```python
def hypothesis(theta,x):
    return theta[0]+theta[1]*x
#h(x)=0+01*x

def error(X,Y,theta):
    total_error=0
    m=X.shape[0]
    
    for i in range(m):
        total_error += (Y[i]-hypothesis(theta,X[i]))**2
        
    return 0.5*total_error 


def gradient(X,Y,theta):
    grad=np.array([0.0,0.0])
    m=X.shape[0]
    for i in range(m):
        grad[0]+= -1*(Y[i]-hypothesis(theta,X[i]))
        grad[1]+= -1*(Y[i]-hypothesis(theta,X[i]))*X[i]
    return grad
                  
def gradientDescent(X,Y,learning_rate,maxitr):
    grad=np.array([0.0,0.0])
    theta=np.array([0.0,0.0])
    e=[]
    
    
    for i in  range(maxitr):
        grad=gradient(X,Y,theta)
        ce=error(X,Y,theta)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        e.append(ce)
    return theta,e
         
theta,e=gradientDescent(X,Y,0.001,100)
                      
print(theta[0],theta[1])
```

```python
plt.scatter(X,Y)
plt.plot(X,hypothesis(theta,X),color='red')
plt.show()

plt.figure(1)
plt.plot(e)
plt.show()
print(e[50],e[99],e[80])


print(hypothesis(theta,9.0))

```
